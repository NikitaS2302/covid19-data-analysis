```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("C:\\Users\\KIIT\\Desktop\\tutedude\\DataScience\\data science\\covid19_Confirmed_dataset.csv")
data.head()
```

```python
df=data.drop(["Lat","Long"],axis=1,inplace=True)
data.head()
```

```python
#aggregated dataset
corona_dataset_agg=data.groupby('Country/Region').sum()
corona_dataset_agg.head()
```

```python
corona_dataset_agg.shape
```

```python
corona_dataset_aggregate=corona_dataset_agg.drop(['Province/State'],axis=1)
corona_dataset_aggregate.head()
```

```python
corona_dataset_aggregate.shape
```

```python
#visualize data according to countries
corona_dataset_aggregate.loc['China'].plot()
corona_dataset_aggregate.loc['India'].plot()
corona_dataset_aggregate.loc['Spain'].plot()
plt.legend()
plt.show()
```

```python
#calculate a good measure
corona_dataset_aggregate.loc['China'].plot()
plt.show()
print('Covid-19 in first 3 days:\n ')
corona_dataset_aggregate.loc['China'][:3].plot()
plt.show()
```

```python
#calculate first derivative
corona_dataset_aggregate.loc['China'].diff().plot()
```

```python
#maximum infection rate
print('Maximum Infection Rate:\n')

print('CHINA: ',corona_dataset_aggregate.loc['China'].diff().max())
print('INDIA: ',corona_dataset_aggregate.loc['India'].diff().max())
print('Spain: ',corona_dataset_aggregate.loc['Spain'].diff().max())
```

```python
countries=list(corona_dataset_aggregate.index)
max_infection_rate=[]
for c in countries:
    max_infection_rate.append(corona_dataset_aggregate.loc[c].diff().max())
corona_dataset_aggregate['Maximum_infection_rate']=max_infection_rate
corona_dataset_aggregate
```

```python
#create a new dataframe
corona_dataset=pd.DataFrame(corona_dataset_aggregate['Maximum_infection_rate'])
corona_dataset
```

```python
happiness_report=pd.read_csv("C:\\Users\\KIIT\\Desktop\\tutedude\\DataScience\\data science\\worldwide_happiness_report.csv")
happiness_report.head()
```
